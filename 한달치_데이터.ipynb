```python
new_data = pd.read_csv('./data/new_data_0520.csv', encoding='utf-8')
new_data = new_data.drop(new_data.columns[0], axis=1)
```

```python
X = new_data.loc[:60000, 'temperature':]
X
```

```python
X = np.array(X)
X.shape
```

```python
temp = []
for i in range(len(X)-360):
  for j in range(360):
    temp.append(X[i+j])
test_data = np.array(temp)
test_data.shape
```

```python
X_test = test_data.reshape(59641,360,10)
X_test.shape
```

```python
ans_0_60000 = model_rnn.predict(X_test)
ans_0_60000 = pd.DataFrame(ans_0_60000)
ans_0_60000
```

```python
for i in range(len(ans_0_60000)):
  if ans_0_60000.iloc[i, 0] < 1:
    ans_0_60000.iloc[i,0] = 0
ans_0_60000
```

```python
ans_0_60000.to_csv('./data/ans_0_60000.csv')
```
